```python
# TRYING TO REALIGN THE DATA
def give_outer_peakind(col,w=np.arange(1,10),limit=0,distance=100):
    npar = col.to_numpy()
    max_peakind = signal.find_peaks_cwt(npar, w)
    max_peakval = npar[max_peakind]
    limit = max(limit, 0)
    max_ = np.argwhere(max_peakval > limit)
    
    # SECOND PEAK
    ind_first_max = max_peakind[max_[0]][0]
    
    i = 1
    d = abs(max_peakind[max_[0]][0] - ind_first_max)

    while(d <= distance):
        ind_first_max = max_peakind[max_[i]][0]
        i += 1
        d = abs(max_peakind[max_[0]][0] - ind_first_max)
   
    d = abs(max_peakind[max_[i]][0] - ind_first_max)
    
    while(d <= distance):
        if max_peakval[max_[i]][0] >= npar[ind_first_max]:
            ind_first_max = max_peakind[max_[i]][0]
            i += 1
            d = abs(max_peakind[max_[i]][0] - ind_first_max)
        else:
            break
    
    # SECOND TO LAST PEAK
    ind_last_max = max_peakind[max_[-1]][0]
    
    i = -2
    d = abs(max_peakind[max_[-1]][0]- ind_last_max)

    while(d <= distance):
        ind_last_max = max_peakind[max_[i]][0]
        i -= 1
        d = abs(max_peakind[max_[-1]][0] - ind_last_max)
    
    d = abs(max_peakind[max_[i]][0] - ind_last_max)

    while(d <= distance):
        if max_peakval[max_[i]][0] >= npar[ind_last_max]:
            ind_last_max = max_peakind[max_[i]][0]
            i -= 1
            d = abs(max_peakind[max_[i]][0] - ind_last_max)
        else:
            break

    return ind_first_max, ind_last_max

# TRYING TO REALIGN THE DATA
def give_outer_peakind_oldfix(col,w=np.arange(1,10),limit=0.025,distance=100):
    npar = col.to_numpy()
    max_peakind = signal.find_peaks_cwt(npar, w)
    max_peakval = npar[max_peakind]
    limit = max(limit, 0)
    max_ = np.argwhere(max_peakval > limit)
    
    
    # FIND THE INNER OUTER PEAKS
    ind_first_max = max_peakind[max_[0]][0]
    ind_last_max = max_peakind[max_[-1]][0]

    i = 1
    d = abs(max_peakind[max_[i]][0] - ind_first_max)
    
    while(d <= distance):
        if max_peakval[max_[i]][0] >= npar[ind_first_max]:
            ind_first_max = max_peakind[max_[i]][0]
            i += 1
            d = abs(max_peakind[max_[i]][0] - ind_first_max)
        else:
            break

    i = -2
    d = abs(max_peakind[max_[i]][0] - ind_last_max)

    while(d <= distance):
        if max_peakval[max_[i]][0] >= npar[ind_last_max]:
            ind_last_max = max_peakind[max_[i]][0]
            i -= 1
            d = abs(max_peakind[max_[i]][0] - ind_last_max)
        else:
            break
    
    return ind_first_max, ind_last_max
```

```python
# IMPROVED FIND OUTER PEAKS VERSION
def give_outer_peakind_improved(col, p=250):
    npar = col.to_numpy()
    first = np.argmax(npar[:p])
    last = len(npar) - p + first
    return first, last
```

```python
# OVERLAP ALLIGNED DATA
sensor = 1
max_index = 1200
for index, row in meta_df.iterrows():
    DATAPATH = 'train'
    DATASET = 'train_' + str(index)

    filename= os.path.join(DATAPATH, DATASET + '.csv')
    col = pd.read_csv(filename)[str(sensor)]
    i,j = give_outer_peakind_improved(col)
    p = col.iloc[i:j]
    p = p.reset_index(drop=True)
    p.plot()
    if index == max_index:
        break
plt.show()
```
