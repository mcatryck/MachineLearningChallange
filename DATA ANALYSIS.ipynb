```python
# imports:

import numpy as np
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt

#%matplotlib notebook
```

```python
metadata = pd.read_csv("train_meta.csv", sep=',',usecols=[1,2,3,4])
```

```python
metadata.head()
```

```python
performance = "diff"
interrest = "direction"
data = {}
if (performance == "diff"):
    for i in range(len(metadata[interrest])):
        if (metadata.loc[i,'label'],metadata.loc[i,interrest]) not in data.keys():
            data[(metadata.loc[i,'label'],metadata.loc[i,interrest])] = pd.read_csv("train/train_{}.csv".format(i), sep=',',usecols=[1,2,3])
else:
    for i in range(len(metadata[interrest])):
        if (metadata.loc[i,'label'],metadata.loc[i,interrest]) not in data.keys():
            data[(metadata.loc[i,'label'],metadata.loc[i,interrest])] = pd.read_csv("train/train_{}.csv".format(i), sep=',',usecols=[1,2,3])
        else:
            data[(metadata.loc[i,'label'],metadata.loc[i,interrest])].append(pd.read_csv("train/train_{}.csv".format(i), sep=',',usecols=[1,2,3]))
```

```python
#plot all diff:
for i in data.keys():
    print('this is label ', i[0],' with regards to ', interrest,'of value', i[1])
    for j in range(3):
        data[i]['{}'.format(j)].plot()
        plt.show()

    
```
